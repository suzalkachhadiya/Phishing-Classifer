```python
#Dropping the Domain column
data.drop(['Domain'], axis = 1,inplace=True)
```

```python

# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
for i in data.columns:
  print(data[i].value_counts())
```

```python
# data.drop(["https_Domain","Right_Click"],axis=1,inplace=True)
```

```python
# Sepratating & assigning features and target columns to X & y
y = data['Label']
X = data.drop('Label',axis=1)
X.shape, y.shape
```

```python
# Splitting the dataset into train and test sets: 80-20 split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2, random_state = 29)
X_train.shape, X_test.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
ML_Model = []
acc_train = []
acc_test = []

#function to call for storing the results
def storeResults(model, a,b):
  ML_Model.append(model)
  acc_train.append(a)
  acc_test.append(b)
```

## decision tree

```python
# Decision Tree model
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion="gini",max_depth = 5)
tree.fit(X_train, y_train)
```

```python
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
```

```python
acc_train_tree = accuracy_score(y_train,y_train_tree)
acc_test_tree = accuracy_score(y_test,y_test_tree)
print(acc_train_tree,acc_test_tree)
```

```python
storeResults('Decision Tree', acc_train_tree, acc_test_tree)
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini', max_depth=10, n_estimators=150, oob_score=True)
forest.fit(X_train, y_train)
```

```python
y_test_forest = forest.predict(X_test)
y_train_forest = forest.predict(X_train)
```

```python
acc_train_forest = accuracy_score(y_train,y_train_forest)
acc_test_forest = accuracy_score(y_test,y_test_forest)
print(acc_train_forest,acc_test_forest)
```

```python
storeResults('Random Forest', acc_train_forest, acc_test_forest)
```

```python
import matplotlib.pyplot as plt
```

```python
feature_importance = forest.feature_importances_
sorted_idx = np.argsort(feature_importance)
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(X_test.columns)[sorted_idx])
plt.title('Feature Importance')
```

## xgboost

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_test_xgb = xgb.predict(X_test)
y_train_xgb = xgb.predict(X_train)
```

```python
acc_train_xgb = accuracy_score(y_train,y_train_xgb)
acc_test_xgb = accuracy_score(y_test,y_test_xgb)
print(acc_train_xgb,acc_test_xgb)
```

```python
storeResults('XGB', acc_train_xgb, acc_test_xgb)
```

## SVM

```python
from sklearn.svm import SVC

svm = SVC(C=3, gamma='auto', kernel='poly')
svm.fit(X_train, y_train)
```

```python
y_test_svm = svm.predict(X_test)
y_train_svm = svm.predict(X_train)
```

```python
acc_train_svm = accuracy_score(y_train,y_train_svm)
acc_test_svm = accuracy_score(y_test,y_test_svm)
print(acc_train_svm,acc_test_svm)
```

```python
storeResults('SVM', acc_train_svm, acc_test_svm)
```
